```python
#/content/drive/MyDrive/Colab Notebooks/proyecto youtube nicolas 
comurl = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/proyecto youtube nicolas /comentariolikelink1.csv",lineterminator='\n')
comurl = comurl.drop('Unnamed: 0',axis=1) 
comurl = comurl.applymap(str) 
comurl
```

```python
coment = comurl['Comment'].to_list()
videoid = comurl['videoID'].to_list()
likes = comurl['likes'].to_list()
```

```python
numcomen=len(comurl)
finallist = []
for i in range(0,numcomen):
  if i%100==0:
    print("x", end = (''))
  if len(coment[i].split()) > 4 :
    newdf = videoid[i], likes[i], coment[i]
    finallist.append(newdf)
```

```python

finallist1=pd.DataFrame(finallist,columns=['VideoID','Likes','Comentario'])

finallist1.to_csv('datosmulti.csv',mode = 'a')
finallist1
```

```python
coment1 = finallist1['Comentario'].to_list()
videoid1 = finallist1['VideoID'].to_list()
likes1 = finallist1['Likes'].to_list()
```

```python
numcomen=len(finallist)
finallist2 = []
for i in range(0,numcomen):
  if i%100==0:
    print("x", end = (''))
  palabra = coment1[i].split()
  for j in reversed(range(0,len(palabra))):
    if len(palabra[j]) <= 3:
      palabra.pop(j)
  newcoment = ' '.join(palabra)
  if len(newcoment)!=0:
    newdf1 = videoid1[i], likes1[i], newcoment
    finallist2.append(newdf1)
```

```python
finallist2=pd.DataFrame(finallist2,columns=['VideoID','Likes','Comentario'])
finallist2.to_csv('datosmulti2.csv',mode = 'a')
```

```python
coment = finallist2['Comentario'].to_list()
videoid = finallist2['VideoID'].to_list()
likes = finallist2['Likes'].to_list()
```
